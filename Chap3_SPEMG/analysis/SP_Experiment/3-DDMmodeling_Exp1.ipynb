```python
import subprocess
import numpy
import pandas as pd

os.chdir("DDMfiles/")

def simulatingRT(a,zr,v,t0,st0,szr,sv, sub, FC, SAT, SS):
    print('Simulating RT for %s' %sub+str(SS))
    p = subprocess.Popen(["wine construct-samples.exe -a %f -z %f -v %f -t %f\
        -Z %f -V %f -T %f -d 0 -n 1000 -N 1 -p 3 -o %s.lst" %(float(a), float(zr), float(v), 
        float(t0), float(szr), float(sv), float(st0), 'synRT/'+ sub + '_' + str(FC) +\
        '_' + str(SAT) + '_' + str(SS))], stdout= subprocess.PIPE, 
        stderr=subprocess.STDOUT, bufsize=1, shell=True)
    m = p.communicate()
    return True

fit = pd.read_csv('fit_results_RT.csv')
for xx, subj_dat in fit.groupby(['participant']): 
    st0 = subj_dat.loc[subj_dat['parameter']=='st0'].value.values[0]
    sv = subj_dat.loc[subj_dat['parameter']=='sv'].value.values[0]
    for stimulus, stimulus_dat in subj_dat.groupby('stimulus'):
        v = stimulus_dat.loc[stimulus_dat['parameter']=='v'].value.values[0]
        for SAT, SAT_dat in subj_dat.groupby('SAT'):
            a = SAT_dat.loc[SAT_dat['parameter']=='a'].value.values[0]
            szr = SAT_dat.loc[SAT_dat['parameter']=='szr'].value.values[0]
            t0 = SAT_dat.loc[SAT_dat['parameter']=='t0'].value.mean()
            for p_right, p_right_dat in SAT_dat.groupby('p_right'):
                zr = p_right_dat.loc[p_right_dat['parameter']=='zr'].value.mean()
                simulatingRT(a,zr,v,t0,st0,szr,sv, str(xx.split('.')[0]), stimulus, p_right, SAT)
                
                fit = pd.read_csv('fit_results_RT.csv')

os.chdir("../")
```

# Testing for effect on bias using mixed models

```python
import pystan
import pickle
import arviz as az
import patsy
```

Initializing stan code

```python
LME = pystan.StanModel(file='MixedModels/LME.stan')
```

Formatting data

```python
LMEsummary = summary[["zr","participant","stimulus","p_right","SAT"]].copy()
LMEsummary = LMEsummary[LMEsummary.stimulus == "right"] #duplicated zr for stimulus column
del LMEsummary['stimulus']
LMEsummary.participant = LMEsummary.participant.replace(LMEsummary.participant.unique(), np.arange(len(LMEsummary.participant.unique()))+1)
LMEsummary.p_right = LMEsummary.p_right.replace([30, 50, 70], [-0.5, 0, 0.5])
LMEsummary.SAT = LMEsummary.SAT.replace(["accuracy","speed"], [1,0])
```

```python
model_name = 'zr'
fixeff_form = "1 + SAT + p_right + SAT:p_right"#Fixed effects formula
raneff_form = fixeff_form#Random effects formula
fixeff = np.asarray(patsy.dmatrix(fixeff_form, LMEsummary)) #FE design matrix
raneff = np.asarray(patsy.dmatrix(raneff_form, LMEsummary))
prior_intercept = np.asarray([.5,.25])
priors_mu = np.repeat(0,3) #Priors on mu for FE
priors_sigma = np.repeat(.25,3) # priors on sigma for FE
priors_raneff = [0,.25] #Priors on RE
prior_sd = [0, .25] #priors on residual sigma

LME_data = dict(
    N = len(LMEsummary),
    P = fixeff.shape[-1], #number of pop level effects
    J = len(LMEsummary.participant.unique()),
    n_u = raneff.shape[-1],
    subj = LMEsummary.participant,
    X = fixeff,
    Z_u = raneff,
    y = LMEsummary.zr.values,
    p_intercept = prior_intercept, p_sd = prior_sd, p_fmu = priors_mu, p_fsigma = priors_sigma, p_r = priors_raneff,
    logT = 0
)

if os.path.exists("MixedModels/FittedModels/%s.pkl"%model_name):
    with open("MixedModels/FittedModels/%s.pkl"%model_name, "rb") as f:
        data_dict = pickle.load(f)
    fit = data_dict['fit']

else:
    fit = LME.sampling(data=LME_data, iter=2000, chains=6, n_jobs=6, 
                              warmup = 1000,thin=2)
    with open("MixedModels/FittedModels/%s.pkl"%model_name, "wb") as f:
        pickle.dump({'model' : LME, 'fit' : fit}, f, protocol=-1)

zr_fit_az = az.from_pystan(posterior=fit, posterior_predictive='y_hat', observed_data="y", log_likelihood='log_lik',
                                       coords={'b': fixeff_form.split('+')[1:]}, dims={'beta': ['b']})
```
