```python
def quantile_cut(dataf, units, split, chron, y, n=7):
    '''
    dataf = dataframe
    units = participants, must be list
    split = factors to spli along (e.g SAT,...), must be list
    chron = chronometric variaqble to make quantiles of
    y = other variable that we want to compute the value at given quantile
    '''
    value, condition, quantile,names, part = [],[],[],[],[]
    tmp = dataf.copy()
    for cell, cell_dat in tmp.groupby(units + split):
        cell_dat = cell_dat.copy()
        cell_dat["quant"] = pd.qcut(cell_dat[chron], n)
        name_quanti = [str(a) for a in cell_dat.groupby("quant")[chron].mean().index]
        mean_chron = cell_dat.groupby("quant")[chron].mean().values
        cell_dat["quant"] = [str(a) for a in cell_dat["quant"]]
        cell_dat["quant"] = cell_dat["quant"].replace(name_quanti, mean_chron)
        cell_dat["names"] = pd.qcut(cell_dat[chron], n, labels=False)
        for quant, quant_dat in cell_dat.groupby('quant'):
            value.append(quant_dat[y].mean())
            condition.append(cell[1:])
            part.append(cell[0])
            quantile.append(quant)
            names.append(quant_dat.names.unique()[0])
    dfout = pd.concat([pd.Series(value), pd.Series(condition),pd.Series(part),pd.Series(quantile),pd.Series(names)], axis=1)
    dfout.columns = ['y','condition','units','quant','names']
    return dfout
```

## Implementation of Stone 1960

```python
import scipy.stats as stats
from matplotlib import gridspec
```

```python
ntrials = 100

distribution_A = stats.norm(.1,1)
distribution_B = stats.norm(-.1,1)
threshold = 4
trials = []
choices = []
for trial in np.arange(ntrials):
    signal = np.random.choice(['A','B'])
    if signal == "A":
        distribution = distribution_A
    else : 
        distribution = distribution_B
    decision_var = [0]
    while np.abs(decision_var[-1]) < threshold :
        sample = distribution.rvs()
        pA = distribution_A.pdf(sample)
        pB = distribution_B.pdf(sample)
        cx = np.log(pA) - np.log(pB)
        decision_var.append(decision_var[-1] + cx)
    trials.append([decision_var,signal])
```

```python
from __future__ import division
import numpy as np
import matplotlib.pyplot as plt

from mpl_toolkits.axes_grid1.inset_locator import inset_axes


fig, ax = plt.subplots(1, 1, figsize=(9, 5), dpi=300)

p_u = 0.4#drift rate (units/sec)
p_a = p_u/7  #upper bound (correct answer)
p_b = -p_a #lower bound (incorrect answer)
p_z = 0
p_s = p_u/3  #standard deviation of drift (units/sec)
t = 0.01

p_dt = .003 #step size for simulations (seconds)

axScatter = ax
axScatter.spines['right'].set_visible(False)
axScatter.spines['top'].set_visible(False)
axScatter.set_yticklabels([])

kde_top = inset_axes(ax, width="100%", height="100%",
                    bbox_to_anchor=(200, p_a+.005, 400, p_a+.01),
                    bbox_transform=ax.transData, borderpad=0)

kde_bottom = inset_axes(ax, width="100%", height="100%",
                    bbox_to_anchor=(200, p_b-.01, 400, p_b-.005),
                    bbox_transform=ax.transData, loc=2, borderpad=0)



kde_top.spines['top'].set_visible(False)
kde_top.spines['bottom'].set_visible(False)
kde_top.spines['left'].set_visible(True)
kde_top.spines['right'].set_visible(False)
kde_top.tick_params(labelleft=False, labelbottom=False)
kde_top.set_xticks([])
kde_top.set_yticks([])

kde_bottom.spines['top'].set_visible(False)
kde_bottom.spines['bottom'].set_visible(False)
kde_bottom.spines['left'].set_visible(True)
kde_bottom.spines['right'].set_visible(False)
kde_bottom.tick_params(labelleft=False, labelbottom=False)
kde_bottom.set_xticks([])
kde_bottom.set_yticks([])

nReps = 500 #number of paths per simulation
nSteps = 600 #number of time steps

dy = p_u*p_dt + p_s*np.sqrt(p_dt)*np.random.randn(nReps, nSteps)

y = []
for i in dy:
    y.append(np.cumsum(i))
y = np.array(y)    
y = y +p_z

t = np.arange (p_dt*1000, (p_dt*nSteps+p_dt)*1000,p_dt*1000)



#plot the bounds
counter = 0
l = 0
m = 0
RT_Cor = []
RT_Inc = []

for i in y:
    first_passage_up = np.where(i >= p_a)[0][0]
    try : 
        first_passage_low = np.where(i <= p_b)[0][0]
    except :
        first_passage_low = 999
    if first_passage_low < first_passage_up:
        k = []
        for j in range(0,nSteps):
            k.append(i[j])
            if i[j] <= p_b :
                k.append(i[j])
                break
        k = k[:j]
        t_ = t[:j]
        censor = np.where(t_ > 400)[0]
        if censor.size == 0:
            censor = -1
        else : 
            censor = censor[0]
        axScatter.plot(t_[:censor]+200,k[:censor] , 'indianred', zorder = 2, alpha=.2,lw=.5)
        RT_Inc.append(t_[-1]+200)
        l += 1
    else :
        k = []
        for j in range(0,nSteps):
            k.append(i[j])
            if i[j] >= p_a :
                k.append(i[j])
                break
        k = k[:j]
        t_ = t[:j]
        censor = np.where(t_ > 400)[0]
        if censor.size == 0:
            censor = -1
        else : 
            censor = censor[0]
        axScatter.plot(t_[:censor]+200,k[:censor] , 'royalblue', zorder = 2, alpha=0.2,lw=.5)
        RT_Cor.append(t_[-1]+200)
        m +=1
    counter = counter + 1


axScatter.set_xlabel(u'Time from stimulus onset', fontsize =17 , fontname="Times New Roman Bold")
axScatter.set_ylabel ('Accumulated Evidence', fontsize = 17)
axScatter.hlines(p_a, 200, nSteps, 'k',zorder = 3, linestyle='dotted', linewidth = 2)
axScatter.hlines(p_b, 200,nSteps, 'k', zorder = 3,linestyle='dotted',  linewidth = 2)
axScatter.hlines(200, 600, nSteps, 'grey', zorder = 3, linewidth = 2)
#axScatter.annotate("", color='b', size=20, xy=(nSteps/5, 0.04), xycoords='data', xytext=(0, 0), textcoords='data', arrowprops=dict(arrowstyle="simple", connectionstyle="arc3", facecolor='b', edgecolor='k'))
#axScatter.annotate("", color='k', size=20, xy=(-10,0), xycoords='data', xytext=(0, 0), textcoords='data', arrowprops=dict(arrowstyle="simple", connectionstyle="arc3", facecolor='b', edgecolor='k'))
#trans = axScatter.get_xaxis_transform() # x in data untis, y in axes fraction
#ann = axScatter.annotate('Encoding Time', xy=(0, 0), xycoords=trans, xytext=(-300, 0), textcoords='data', arrowprops=dict(arrowstyle="simple", connectionstyle="arc3", facecolor='b', edgecolor='k'))
#axScatter.text(30, 0.029, "v", color='w', fontsize=38, rotation=28)
#axScatter.text(30, 0.030, "v", color='b', fontsize=35, rotation=28)
axScatter.vlines(200, p_a, p_b,  'k',zorder = 3, linewidth = 2)
axScatter.vlines(600, p_a, p_b,  'k',zorder = 3, linewidth = 2)
axScatter.text(25, -0.02, '$\mathit{T_{encoding}}$', color='k', fontsize=20)
axScatter.text(610, -0.02, '$\mathit{T_{response}}$', color='k', fontsize=20)
    #axScatter.hlines(0, 600,900, 'k', zorder = 3, linestyle='dashed', linewidth = 2)
axScatter.annotate("", color='k', size=10, xy=(180,0), xycoords='data', xytext=(0, 0), textcoords='data', arrowprops=dict(arrowstyle="simple", connectionstyle="arc3", facecolor='k', edgecolor='k'))
axScatter.annotate("", color='k', size=10, xy=(740,0), xycoords='data', xytext=(600, 0), textcoords='data', arrowprops=dict(arrowstyle="simple", connectionstyle="arc3", facecolor='k', edgecolor='k'))
axScatter.set_yticks(np.arange (-0.1, 0.11, 0.05))

axScatter.set_xlim(0, 800)
#axScatter.text(205, p_a+0.005,u'Upper Threshold',fontsize=10,fontname="Times New Roman Bold",color="grey")
#axScatter.text(205, p_b-0.007,u'Lower Threshold',fontsize=10,fontname="Times New Roman Bold",color="grey")


axScatter.text(150, p_a-0.005,u'$\mathit{a+}$',fontsize=20,fontname="Times New Roman Bold")
axScatter.text(150, p_b-0.005,u'$\mathit{a-}$',fontsize=20,fontname="Times New Roman Bold")
axScatter.text(180, p_z-0.005,u'$\mathit{z}$',fontsize=20,fontname="Times New Roman Bold")
axScatter.hlines(0, 200,600, ls='--',color='k')

axScatter.annotate("", color='k', size=20, xy=(200+150, p_u*.150), xycoords='data', xytext=(200, p_z), textcoords='data', arrowprops=dict(arrowstyle="simple", connectionstyle="arc3", facecolor='k', edgecolor='k'))
axScatter.text(215, p_a/3, r"$\mathit{v}$", color='k', fontsize=20, rotation=28)
#axScatter.text(300, 0.10,r'$dx = \mathcal{A}dt + cd\mathcal{W}$',fontsize=25,fontname="Times New Roman Bold")
#axScatter.text(250, 0.10,r'${TR} = T_{Decision} + T_{er}$',fontsize=25,fontname="Times New Roman Bold")
#axScatter.text(300, -0.12,r'$\overline{RC} = \frac{1}{1 + e^{-2A\mathit{k|x|}}}$',fontsize=30,fontname="Times New Roman Bold")
axScatter.xaxis.set_ticks(np.arange(0, 10000, 10000))
axScatter.set_ylim(-.14,.14)

kde_top.set_xlim(200, 600)
kde_bottom.set_xlim(200, 600)
sns.kdeplot(RT_Cor, ax=kde_top, color='royalblue',bw=.25, cut=0)
sns.kdeplot(RT_Inc, ax=kde_bottom, color='indianred',bw=.25, cut=0)

#plt.tight_layout()
#plt.savefig('DDM.png')
```
