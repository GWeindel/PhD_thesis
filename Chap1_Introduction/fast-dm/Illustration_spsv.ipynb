```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import subprocess
import os
```

# Creating dataset

```python
os.chdir('DDMfiles/')
def simulatingRT(a,zr,v,t0,st0,szr,sv, sub):
    print("wine construct-samples.exe -a %f -z %f -v %f -t %f\
        -Z %f -V %f -T %f -d 0 -n 3000 -N 1 -p 3 -o %s.lst" %(float(a), float(zr), float(v), 
        float(t0), float(szr),float(sv),float(st0), 'gen_synRT/'+ sub))
    p = subprocess.Popen(["wine construct-samples.exe -a %f -z %f -v %f -t %f\
        -Z %f -V %f -T %f -d 0 -n 3000 -N 1 -p 3 -r -o %s.lst" %(float(a), float(zr), float(v), 
        float(t0), float(szr),float(sv),float(st0), 'gen_synRT/'+ sub)], stdout= subprocess.PIPE, 
        stderr=subprocess.STDOUT, bufsize=1, shell=True)
    m = p.communicate()
    return True

a,v,t0,zr,szr,sv,st0 = .85, 1, 0.35, 0.5, 0.8,.3,0
for i in np.arange(4)+1:
    v = i - 0.5
    simulatingRT(a,zr,v,t0,st0,szr,sv, str(i))
os.chdir('../')
```

```python

list_dir = os.listdir('DDMfiles/gen_synRT/')
list_ = []
for f in list_dir:
    raw_ = pd.read_csv('DDMfiles/gen_synRT/'+f,header=None,sep="  ")
    raw_['stimulus'] = f.split('.')[0]
    raw_['participant'] = "S0"
    list_.append(raw_)
df = pd.concat(list_, ignore_index=True)
df.columns= ['response','RT','stimulus','participant']
```

## Fitting Model on RT

Running script DDMfiles/DDMfit.py to generate estimated parameters (extremely slow, cell after this one collects results)

```python
import glob,os
precision = 3
method = 'ks'#You can switch with ml, cs or ks

Free = {
    'v':['stimulus']}

Fixed = {'p':0,
         'd':0,
         'szr':0,
         'sv':0,
         'st0':0
          }

os.chdir('DDMfiles/')
import FDM_functions_SATTM as fdm
fdm.generate_files(df, free = Free, fixed = Fixed, method=method, precision=precision)

fdm.fit(nproc=1)

pars = fdm.get_parameters()

pars.to_csv('fit_results_RT_wovar.csv')

for f in glob.glob("*.ctl"):
    os.remove(f)
    
for f in glob.glob("*.lst"):
    os.remove(f)
    
for f in glob.glob("data_*.csv"):
    os.remove(f)
os.chdir('../')
```

```python
os.chdir('DDMfiles/')
def simulatingRT(a,zr,v,t0,st0,szr,sv, sub):
    print("wine construct-samples.exe -a %f -z %f -v %f -t %f\
        -Z %f -V %f -T %f -d 0 -n 3000 -N 1 -p 3 -o %s.lst" %(float(a), float(zr), float(v), 
        float(t0), float(szr),float(sv),float(st0), 'recov_synRT/'+ sub))
    p = subprocess.Popen(["wine construct-samples.exe -a %f -z %f -v %f -t %f\
        -Z %f -V %f -T %f -d 0 -n 3000 -N 1 -p 3 -o %s.lst" %(float(a), float(zr), float(v), 
        float(t0), float(szr),float(sv),float(st0), 'recov_synRT/'+ sub)], stdout= subprocess.PIPE, 
        stderr=subprocess.STDOUT, bufsize=1, shell=True)
    m = p.communicate()
    return True

fit = pd.read_csv('fit_results_RT_wovar.csv')

for xx, subj_dat in fit.groupby(['participant']): 
    zr = subj_dat.loc[subj_dat['parameter']=='zr'].value.mean()
    a = subj_dat.loc[subj_dat['parameter']=='a'].value.values[0]
    t0 = subj_dat.loc[subj_dat['parameter']=='t0'].value.values[0]
    st0,sv,szr = 0,0,0
    for contrastXstim, c_dat in subj_dat.groupby(['stimulus']):
        v = c_dat.loc[c_dat['parameter']=='v'].value.values[0]
        simulatingRT(a,zr,v,t0,st0,szr,sv, str(contrastXstim))
os.chdir('../')
```

```python
list_dir = os.listdir('DDMfiles/recov_synRT/')
list_ = []
for f in list_dir:
    raw_ = pd.read_csv('DDMfiles/recov_synRT/'+f,header=None,sep="  ")
    raw_['stimulus'] = f.split('.')[0]
    raw_['participant'] = "S0"
    list_.append(raw_)
recovdf = pd.concat(list_, ignore_index=True)
recovdf.columns= ['response','RT','stimulus','participant']
```

```python
plt.figure(dpi=300, figsize=(4,4))
name = "simple DDM"
syndata = recovdf

Prec, RTQuantiles, subject, contrast = [],[],[],[]
meanPrec, meanRT,synmeanPrec, synmeanRT = [],[],[],[]
meanRT_sub, meanPrec_sub,synmeanRT_sub, synmeanPrec_sub = [],[],[],[]
for participant, part_dat in df.groupby('participant'):    
    con_meanRT, con_meanPrec,syncon_meanRT, syncon_meanPrec = [],[],[],[]
    for con, con_dat in df.groupby("stimulus"):
        corr_meanRT, corr_meanPrec,syncorr_meanRT, syncorr_meanPrec = [],[],[],[]
        for corr, corr_dat in con_dat.groupby("response"):
            corr_meanRT.append(corr_dat.RT.quantile([.1,.3,.5,.6,.9]).values*1000)
            corr_meanPrec.append(np.repeat(float(len(corr_dat.response))/len(con_dat),5))
            syn_subset = syndata[(syndata.stimulus == con)]
            syncorr_meanRT.append(syn_subset[syn_subset.response == corr].RT.quantile([.1, .3,.5,.6,.9])*1000)
            syncorr_meanPrec.append(np.repeat(syn_subset[syn_subset.response == corr].response.count()/syn_subset.response.count(),5))
        con_meanRT.append(corr_meanRT)
        con_meanPrec.append(corr_meanPrec)
        syncon_meanRT.append(syncorr_meanRT)
        syncon_meanPrec.append(syncorr_meanPrec)
    meanRT_sub.append(con_meanRT)
    meanPrec_sub.append(con_meanPrec)
    synmeanRT_sub.append(syncon_meanRT)
    synmeanPrec_sub.append(syncon_meanPrec)

meanRT.append([np.mean(meanRT_sub, axis=0)])
meanPrec.append([np.mean(meanPrec_sub, axis=0)])
meanRT, meanPrec = np.asarray(meanRT[0][0]), np.asarray(meanPrec[0][0])

synmeanRT.append([np.mean(synmeanRT_sub, axis=0)])
synmeanPrec.append([np.mean(synmeanPrec_sub, axis=0)])
synmeanRT, synmeanPrec = np.asarray(synmeanRT[0][0]),np.asarray(synmeanPrec[0][0])


curax = plt

ys = synmeanRT
colors =  ['indianred','royalblue']
i, col = 0,0
for y in ys:
    for y_i in np.arange(2):
        curax.plot(synmeanPrec[i,y_i].flatten(), y[y_i].flatten(), marker="o",  color=colors[y_i], alpha=1/(i+1))
        curax.plot(meanPrec[i,y_i].flatten(),meanRT[i,y_i].flatten(), 'x', color=colors[y_i], markeredgecolor="k")
    i += 1
    
curax.xlabel("Response proportion")
curax.xlim(0,1)
curax.vlines(.5,0,2000,linestyle=':')

curax.annotate('Correct', xy=(.9, 725), xytext=(.52,725),
             arrowprops={'arrowstyle': '->'}, va='center')
curax.annotate('Errors', xy=(.1, 725), xytext=(.35,725),
             arrowprops={'arrowstyle': '->'}, va='center')
curax.plot( [0,0],  [0,0],'x', label="Data", color='k')
curax.plot( [0,0],  [0,0],'o', label="Model", color='gray', markerfacecolor="w", markeredgecolor="gray", alpha=.2)
curax.legend(loc='upper right', bbox_to_anchor=(1.02, .97))
curax.ylim([250, 750])
plt.ylabel('Mean RT of quantile (ms)')

plt.tight_layout()
plt.savefig('DDMfiles/QPplot/%s.png'%name)
plt.show() 
```
